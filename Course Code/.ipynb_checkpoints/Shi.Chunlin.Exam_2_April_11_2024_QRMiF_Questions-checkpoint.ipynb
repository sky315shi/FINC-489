Question 2: 10 points) Download 48 industry portfolios, '48_Industry_Portfolios'. Make excess returns of industry porfolios by computing industry returns minus the risk-free rate ('RF'). 'RF' is available in the Fama-French 5 factor data you made in question 1.

```python
# Code here
ds2 = web.DataReader('48_Industry_Portfolios', 'famafrench', start=sdate, end=edate)[0]
ds2
```

```python
ds2_minus_RF = ds2.sub(ds1['RF'], axis=0)
ds2_minus_RF
```

Bootsrap function for mean is given as below, but if you prefer, you can build your own codes.

```python
# Bootstrap function for mean (time-series average for fund return minus SPY)

def bootstrap_mean(data, n_bootstrap=1000):
    bootstrap_means = []
    for _ in range(n_bootstrap):
        bootstrap_sample = np.random.choice(data[~np.isnan(data)], size=len(data), replace=True) #sampling with replacement for observed data
        bootstrap_means.append(np.mean(bootstrap_sample)) # Each repeat will add a mean based on newly bootstrapped samples
    return bootstrap_means
```

Question 3: 20 points) Run bootstrap tests to compute t-statistics and confidence interval for the excess returns of industries.

```python
# Code here
bootstrap_results = {}
for ticker in ds2_minus_RF.columns:
    bootstrap_means = bootstrap_mean(ds2_minus_RF[ticker])
    mean_estimate = np.mean(bootstrap_means)
    std_error = np.std(bootstrap_means, ddof=1) #standard error = std dev of estimate (i.e., mean in this case)
    t_statistic = mean_estimate / std_error
    conf_interval = stats.norm.interval(0.95, loc=mean_estimate, scale=std_error)
    bootstrap_results[ticker] = {'t_statistic': t_statistic, 'conf_interval': conf_interval}
bootstrap_results
```

Question 4: 20 points) Sort the boostrap results by the size of the t-statistics and show the results. Which industry has the highest positive t-stats? How about the lowest t-stat industry? Explain your answer

```python
# Code here
# Convert to dataframe for easier manipulation
for industry in bootstrap_results:
    bootstrap_results[industry]['CI Lower'], bootstrap_results[industry]['CI Upper'] = bootstrap_results[industry].pop('conf_interval')
df_bootstrap_results = pd.DataFrame.from_dict(bootstrap_results, orient='index')
df_bootstrap_results
```

```python
# Sort the boostrap results by the size of the t-statistics
df_bootstrap_results_sorted = df_bootstrap_results.sort_values(by='t_statistic')
df_bootstrap_results_sorted
```

From above, we can see that the industry with highest t-statistic is Smoke, and the industry with the lowest t-statistic is Health. This implies that Smoke industry's excess returns are significantly higher than the risk-free rate and Health industry implies that its mean excess returns are significantly lower than the risk-free rate. Hence, Smoke industry had a strong performance over the period studied, yielding returns that consistently outperformed the risk-free rate. While for Health industry, it underperformed relative to the risk-free rate during the study period or that it carries lower risk. The reasons may because Smoke industry yields a higher profit or higher risks, while health industry yields a lower profit or lower risks.

Question 5: 25 points) Construct the highest t-stat industry minus the lowest t-stat industry as a zero-cost return portfolio. Then, compare between this portfolio and Fama-French factors by plotting scatter plots.

```python
# Code here
ds2['Smoke_minus_Hlth'] = ds2['Smoke'] - ds2['Hlth ']
factors = pd.concat([ds1, ds2.Smoke_minus_Hlth], axis=1)
factors
```

```python
column_names = ['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA']

for i in column_names:
    plt.figure(figsize=(10, 6))
    plt.scatter(factors[i], factors['Smoke_minus_Hlth'], alpha=0.7)
    plt.title(f'Scatter Plot of Smoke_minus_Hlth vs. {i}')
    plt.xlabel(i)
    plt.ylabel('Excess Returns of Smoke_minus_Hlth')
    plt.grid(True)
    plt.show()
```
