```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas_datareader.data as web
yf.pdr_override()
```

Q1: 10 points) Download the following price data from yahoo finance (S&P500, S&P500 Futures, and Treasury 3-month interest rate) between Jan 2, 2000 and Mar 23, 2024. Their tickers are ('^GSPC', 'ES=F', '^IRX'), respectively.

```python
# Code here
tickers = ['^GSPC', 'ES=F', '^IRX']
start_date = '2000-01-02'
end_date = '2024-03-23'
data = yf.download(tickers, start=start_date, end=end_date)
```

```python
data
```

Q2: 10 points) Create a data copy and extract Adjusted Close Price ('Adj Close') from the data set you downloaded. Show the result.

```python
# Data copy
dcopy = data.copy()
dcopy
```

```python
# Extracting Adjusted Close Price
data_AdjP = dcopy['Adj Close']
data_AdjP
```

Q3: 10 points) Compute the daily return of S&P500 index between $t$ and $t+1$ to add as a column to the data frame you created in Question 2. (You can use 'pct_change' commmand.)

```python
data_AdjP['SP500_ret_1'] = data_AdjP["^GSPC"].pct_change(periods=1)
data_AdjP
```

Q4: 10 points) Compute the daily return of S&P500 index between  𝑡  and  𝑡+21 days  to add as a column to the data frame you created in Question 3. (You can use 'pct_change' commmand.

```python
data_AdjP['SP500_ret_21'] = data_AdjP["^GSPC"].pct_change(periods=21)
data_AdjP
```

Q5: 20 points) Compute excess future returns by subtracting Treasury interest rates from the returns computed. You need to adjust the interest rate which is expressed in percentage by dividing by 100 and adjust for daily returns. That is, if your data set is dataAdjP and the name of the interest rate is '^IRX', then you need to make the following transformation: data_AdjP['^IRX'] / 100 / 365 in case of the daily return. For the monthly, you want to adjust by data_AdjP['^IRX'] / 100 * 30 / 365

```python
# For the Treasury 3-month interest rate for daily basis
daily_treasury_rate = data_AdjP['^IRX'] / 100 / 365
daily_treasury_rate
```

```python
# For 21-day basis
treasury_rate_21day = data_AdjP['^IRX'] / 100 * 21 / 365
treasury_rate_21day
```

```python
data_AdjP['Excess Daily Return'] = data_AdjP['SP500_ret_1']-daily_treasury_rate
data_AdjP['Excess 21-day Return'] = data_AdjP['SP500_ret_21']-treasury_rate_21day
data_AdjP
```

Q6: 20 points) Create a variable called future spot spread ratio by computing Futures price minus stock price and then divide it by the stock price. Add it to your data. Make sure to use dropna at the end so that it does not have a missing value in your final data set.

```python
data_AdjP = data_AdjP.dropna()
data_AdjP['Future Spot Spread Ratio'] = (data_AdjP['ES=F'] - data_AdjP['^GSPC']) / data_AdjP['^GSPC']
data_AdjP
```

Q7: 10 points) Plot the relations between future excess returns and the futures-spot spread. Interpret your results.

```python
# We first calculate daily future excess returns
data_AdjP['SP500_future_ret_1'] = data_AdjP["ES=F"].pct_change(periods=1)
data_AdjP['Future Excess Daily Return'] = data_AdjP['SP500_future_ret_1']-daily_treasury_rate
data_AdjP
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_AdjP['Future Spot Spread Ratio'], data_AdjP['Future Excess Daily Return'], alpha=0.5)
plt.title('Relationship between Future Excess Daily Returns and Futures-Spot Spread Ratio')
plt.xlabel('Futures-Spot Spread Ratio')
plt.ylabel('Future Excess Daily Returns')
plt.grid(True)
plt.show()
```

From above, we can see that there exist a slightly positive relationship between Future Excess Daily Return and Future Spot Spread Ratio 

Q8: 10 points + Bonus 5 points) Can you run a predictive regression between the excess returns and the future spot spread ratio? Interpret your results.

```python
data_AdjP = data_AdjP.dropna()
X = data_AdjP['Future Spot Spread Ratio']
y = data_AdjP['Future Excess Daily Return']
X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
results.summary()
```

From above, the coefficient for the "Future Spot Spread Ratio" is 0.0920, which suggests a positive relationship between the spread ratio and future excess daily returns.

The p-value for the "Future Spot Spread Ratio" is 0.044, which indicates that the coefficient is statistically significant, and there is evidence to suggest that the spread ratio does predict future excess daily returns.

The R-squared value is 0.001, which suggests that while the spread ratio may be significant, it is not a strong predictor of future returns.
