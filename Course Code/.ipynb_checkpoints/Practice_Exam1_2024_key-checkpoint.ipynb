Q1) Download the following price data from yahoo finance (S&P500, VIX, VIXVOL) between Jan 2, 2007 and Feb 23, 2024. Their tickers are (^GSPC, ^VIX, ^VVIX), respectively.

```python
Ticker = ["^GSPC","^VIX","^VVIX"]
data = web.get_data_yahoo(Ticker, start="2007-01-02", end="2024-02-23")
```

```python
data
```

Question 2) Create a data copy and extract Adjusted Close Price ('Adj Close') from the data set you downloaded. Show the result.

```python
dcopy = data.copy() # Create a copy to work with
data_AdjP = dcopy['Adj Close'] # Just pick adjusted close date for each ticker
data_AdjP =data_AdjP.dropna() # Remove NaN observations (Not necessary)
data_AdjP # Show the results for your review
```

```python
data_AdjP['RetSP500_nxt'] = data_AdjP["^GSPC"].pct_change(periods=-1) #Compute the next period daily return
```

```python
data_AdjP['RetSP500_t'] = data_AdjP["^GSPC"].pct_change(periods=1) #Compute the current period daily return
```

```python
data_AdjP['RetSP500_nxtLT1'] = data_AdjP["^GSPC"].pct_change(periods=-21*1) #Compute the next 21-day ahead daily return
```

```python
data_AdjP['RetSP500_nxtLT2'] = data_AdjP["^GSPC"].pct_change(periods=-21*2) #Compute the next 42-day ahead daily return
```

```python
data_AdjP['RetSP500_nxtLT3'] = data_AdjP["^GSPC"].pct_change(periods=-21*3) #Compute the next 63-day ahead daily return
```

```python
data_AdjP = data_AdjP.dropna()
```

```python
data_AdjP
```
